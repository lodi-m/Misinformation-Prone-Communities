Possible Ways to get final training dataset: 
1.  - Take a random sample of 10,000 from real tweets, real retweets and real news headlines 
      - i.e randomly taking a sample of 10,000 from ~17,000 real tweets, ~10,000 retweets and ~7,000 real news headlines
    - Use all the fake tweets, the fake retweets and randomly choosing the remaining entries from the fake news headlines

2.   - Take a random sample of 10,000 from just the real tweets and real retweets

I will be trying number 1 first, and depending on the base run and it's accuracy, may choose to use number 2 afterwards.

Data cleaning


```python
real_data = pd.read_csv("/content/real_coaid_comb.csv")
fake_tweets = pd.read_csv("/content/fake_tweets_coaid_comb")
fake_news = pd.read_csv("/content/fake_news_coaid_comb")
```

```python
real_data['text'] = real_data['text'].str.lower().str.replace('rt|@[^\s]+|[^A-Za-z ]', '', regex=True)
real_data['text'] = real_data['text'].str.lower().str.replace('http\S+', '', regex=True) 
real_data['text'].duplicated().sum()
```

```python
real_data = real_data.drop_duplicates(subset=['text'])
real_data['text'].duplicated().sum()
```

```python
real_data.reset_index(drop=True)
```

```python
real_data.to_csv('all_clean_real_coaid_data.csv', index=False)
```

```python
fake_tweets['text'] = fake_tweets['text'].str.lower().str.replace('rt|@[^\s]+|[^A-Za-z ]', '', regex=True)
fake_tweets['text'] = fake_tweets['text'].str.lower().str.replace('http\S+', '', regex=True) 
fake_tweets['text'].duplicated().sum()
```

```python
fake_tweets = fake_tweets.drop_duplicates(subset=['text'])
fake_tweets['text'].duplicated().sum()
```

```python
fake_news['text'] = fake_news['text'].str.lower().str.replace('rt|@[^\s]+|[^A-Za-z ]', '', regex=True)
fake_news['text'] = fake_news['text'].str.lower().str.replace('http\S+', '', regex=True) 
fake_news['text'].duplicated().sum()
```

```python
fake_news = fake_news.drop_duplicates(subset=['text'])
fake_news['text'].duplicated().sum()
```

```python
fake_tweets.reset_index(drop=True)
fake_news.reset_index(drop=True)
```

```python
fake_tweets.to_csv('clean_fake_tweets_coaid.csv', index=False)
fake_news.to_csv('clean_fake_news_coaid.csv', index=False)
```
