Possible Ways to get final training dataset: 
1.  - Take a random sample of 10,000 from real tweets, real retweets and real news headlines 
      - i.e randomly taking a sample of 10,000 from ~17,000 real tweets, ~10,000 retweets and ~7,000 real news headlines
    - Use all the fake tweets, the fake retweets and randomly choosing the remaining entries from the fake news headlines

2.   - Take a random sample of 10,000 from just the real tweets and real retweets
    - Use all the fake tweets, the fake retweets and randomly choosing the remaining entries from the fake news headlines

I will be trying number 1 first, and depending on the base run and it's accuracy, may choose to use number 2 afterwards.

Real data csvs

```python
import pandas as pd
```

```python
# appending all the real csvs together and assigning them the proper labels
real_052020_tweets = pd.read_csv("/content/drive/MyDrive/MPC/claim_real_05-01-2020.csv")
real_data = pd.DataFrame(columns=list(real_052020_tweets.columns))

#All the csv files contain the same columns and column names, can just append them as usual
real_tweet_paths = ['/content/drive/MyDrive/MPC/claim_real_05-01-2020.csv', '/content/drive/MyDrive/MPC/claim_real_07-01-2020.csv', \
               '/content/drive/MyDrive/MPC/claim_real_09-01-2020.csv', '/content/drive/MyDrive/MPC/claim_real_11-01-2020.csv', \
               '/content/drive/MyDrive/MPC/rt_real_05-01-2020.csv', '/content/drive/MyDrive/MPC/rt_real_07-01-2020.csv', \
               '/content/drive/MyDrive/MPC/rt_real_09-01-2020.csv', '/content/drive/MyDrive/MPC/rt_real_11-01-2020.csv', \
               '/content/drive/MyDrive/MPC/news_real_05-01-2020.csv', '/content/drive/MyDrive/MPC/news_real_07-01-2020.csv', \
               '/content/drive/MyDrive/MPC/news_real_09-01-2020.csv', '/content/drive/MyDrive/MPC/news_real_11-01-2020.csv']

for path in real_tweet_paths:
  real_data = real_data.append(pd.read_csv(f'{path}'), ignore_index=True, verify_integrity=True)

real_data['label'] = 1
```
